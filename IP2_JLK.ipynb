### Step 4:
The column names are unwieldy for our purposes. This code renames them according to this list {date_time, temp_c,
dew_point, relative_humidity, wind_speed, visibility, pressure, and weather}.

```python
df.rename(columns = {'Date/Time':'date_time', 
                     'Temp (C)':'temp_c', 
                     'Dew Point Temp (C)':'dew_point', 
                     'Rel Hum (%)':'relative_humidity',
                     'Wind Spd (km/h)':'wind_speed', 
                     'Visibility (km)':'visibility', 
                     'Stn Press (kPa)':'pressure', 
                     'Weather':'weather'
                    }, inplace = True)
```

### Step 6:
This code is used to add a column named temp_f to the dataset that contains the temperature from temp_c
represented in Fahrenheit.

```python
df['temp_f'] = df['temp_c'] * 1.8 + 32
df.head()
```

### Step 7:
This code is used to rearrange the columns in the dataset so temp_f appears immediately before temp_c.

```python
mid = df['temp_f']
df.drop(labels = ['temp_f'], axis = 1, inplace = True)
df.insert(1,'temp_f',mid)
df.head(15)
```

### Step 8
This code is used to display the data types of all the columns in the dataset.

```python
df.dtypes
```

### Step 9:
You’ll notice that the date_time column is not actually considered a date by Pandas. This code is used to
convert this column to a date/time column. Verify the conversion was successful.

```python
df['date_time'] = pd.to_datetime(df['date_time'])
df.dtypes
```

### Step 10:
This code is used to get the average temperature for each month and put it into a dataframe (not a series, you
might need to do some conversion) named monthly_avg.

```python
month = df.date_time.dt.month
average = df.groupby(month).temp_f.mean()
monthly_avg = pd.DataFrame({'monthly_avg': average   
                    })
monthly_avg
```

### Step 11:
This code is used promote the index for monthly_avg to a column called ‘month’. This column will contain 1,
2, . . ., 11, 12.

```python
monthly_avg.reset_index(inplace = True)
monthly_avg = monthly_avg.rename(columns={'date_time': 'month'})
monthly_avg
```

### Step 12:
This code is used to reorder the monthly_avg columns so they appear [‘month’, ‘temp_f’]

```python
mid = month
df.insert(1,'month',mid)
df.sample(15)
```

### Step 14:
This code is used to change each entry in the monthly_avg.month column so it shows the name of the month
and instead of the number. You must do this by creating a function that accepts the month (number)
and returns the month name which you call from a map/lambda expression. A dictionary is useful
here.

```python
month_name = { 1: 'January',
               2: 'February',
               3: 'March',
               4: 'April',
               5: 'May',
               6: 'June',
               7: 'July',
               8: 'August',
               9: 'September',
               10: 'October',
               11: 'November',
               12: 'December'
            }

df['month'] = df['month'].apply(lambda x: month_name[x])
df.head()
```

### Step 15:
This code is used to get the average temperature for each hour starting from 00:00:00 to 23:00:00 (you can
leave this in a series – you are going to manipulate the index).

```python
hour = df.date_time.dt.hour
average = df.groupby(hour).temp_f.mean()
hourly_avg = pd.Series(average)
hourly_avg.to_frame()
```

### Step 16:
This code is used to rename the hourly_avg index to ‘Hour’.

```python
hourly_avg.index.name = 'Hour'
hourly_avg.to_frame()
```

### Step 17:
This code is used to change each entry in the hourly_avg index so it shows the name of the standard time (not
military time). You must do this by creating a function that accepts the time (string) and returns the
time in the format (12:00AM, 1:00AM, 2:00AM, . . . 10:00PM, 11:00PM) which you call from a
map/lambda expression. A dictionary is useful here.

```python
hour_name = {  0: '12:00AM',
               1: '1:00AM',
               2: '2:00AM',
               3: '3:00AM',
               4: '4:00AM',
               5: '5:00AM',
               6: '6:00AM',
               7: '7:00AM',
               8: '8:00AM',
               9: '9:00AM',
               10: '10:00AM',
               11: '11:00AM',
               12: '12:00PM',
               13: '1:00PM',
               14: '2:00PM',
               15: '3:00PM',
               16: '4:00PM',
               17: '5:00PM',
               18: '6:00PM',
               19: '7:00PM',
               20: '8:00PM',
               21: '9:00PM',
               22: '10:00PM',
               23: '11:00PM'
             }
hourly_avg.index = list(map(lambda x: hour_name[x], hourly_avg.index))
hourly_avg.to_frame()
```

### Step 19:
You will notice that when you look at the output from the previous code that some of the rows in the
weather column have multiple weather conditions (e.g. Rain, Snow, Fog). This code is used to get a list that
